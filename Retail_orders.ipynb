```python
df = pd.read_csv("python revision/orders.csv",na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
# rename the column name to lower case and replace space with underscore
df.columns
```

```python
df.columns=df.columns.str.lower()
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df.head()
```

```python
# derive a new colums discount, sales price and  profit
df['discount']=df['list_price']*df['discount_percent']*0.01
df.head()
```

```python
df['sale_price']=df['list_price']-df['discount']
df.head()
```

```python
df['Profit']= df['sale_price']-df['cost_price']
df.head()
```

```python
# now cost_price,list_price and discount_percent is not required
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
